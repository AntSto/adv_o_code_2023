```python
import pandas as pd
df = pd.read_csv(
            r'day7.txt',sep=':', encoding='latin-1', header=None)
df.columns=['original']
df=df['original'].str.split(' ', expand =True)
df.columns=['hand', 'bid']
```

```python
order = ['A','K', 'Q', 'J', 'T', '9', '8', '7', '6', '5', '4', '3', '2']
i=0
dic_value={}
while i < len(order):
    dic_value[order[i]]=str(len(order)-i)
    i+=1
# dic_value
```

```python
df['each']=df['hand'].str.split('').str[1:-1] #.str.join(' ')
# on remplace les valeurs de carte par les points (servira plus tard pour les égalités)
df['total'] = df['each'].apply(lambda x: [int(dic_value[a]) if a in dic_value else a for a in x])
df['total_sum']= [sum(a) for a in df['total']]
df
```

```python
# ordonne total pour être sûr de prendre la bonne première carte pour les égalités high to low
df['total_htl']=df["total"].apply(lambda x: sorted(list(x), reverse=True))
#non on doit prendre dans l'ordre!!
```

```python
# faire la liste de toutes les mains (les 5 cartes et pas que les 2 premières)
```

```python
from collections import Counter 
i=0
game= []
valeur = []
card_1=[]
card2=[]
card3=[]
card4=[]
card5=[]
while i<len(df['hand']):
    total=df['total'][i]
#     print( dict(Counter(df['each'][i])))
    card = dict(Counter(df['total'][i]))
    game.append(card)
    # on récupère le mx nb de cartes identique
    max_key, max_same = max(card, key=card.get), card[max(card, key=card.get)]
#     print(card[max(card, key=card.get)])
    if max_same ==5:
        # 5 cartes identiques les autres à 2
        valeur.append(7)  # 7 car 7 possibilité dans l'exemple
        card_1.append(max_key)
        card2.append('2')
        card3.append('2')
        card4.append('2')
        card5.append('2')

    elif max_same ==4:
        # 4 cartes, on del pour la 5e carte
        valeur.append(6)
        card_1.append(total[0])
        # del card[max_key]
        # max_key, max_same = max(card, key=card.get), card[max(card, key=card.get)]
        card2.append(total[1])
        card3.append(total[2])
        card4.append(total[3])
        card5.append(total[4])
    
    # main de 3 same
    elif max_same ==3:
#         print(max(card, key=card.get))
#         print(card, 'mk', card[max_key])
        card_1.append(total[0])
        del card[max_key]
        max_key, max_same = max(card, key=card.get), card[max(card, key=card.get)]
#         print('2', card[max(card, key=card.get)])
        if max_same == 2:
            valeur.append(5)
            card2.append(total[1])
            card3.append(total[2])
            card4.append(total[3])
            card5.append(total[4])
        else:
            # on a 3 cartes identiques et 2 différentes on trie tou
            valeur.append(4)
            card2.append(total[1])
            del card[max_key]
            max_key, max_same = max(card, key=card.get), card[max(card, key=card.get)]
            
            card3.append(total[2])
            card4.append(total[3])
            card5.append(total[4])
#         
    elif max_same ==2:
#         print(max(card, key=card.get))
        card_1.append(total[0])
        del card[max_key]
        max_key, max_same = max(card, key=card.get), card[max(card, key=card.get)]
#         print(df['each'][i])
#         print(max(card, key=card.get))
        if max_same == 2:
            valeur.append(3)
            card2.append(total[1])
            del card[max_key]
            max_key, max_same = max(card, key=card.get), card[max(card, key=card.get)]
            
            card3.append(total[2])
            card4.append(total[3])
            card5.append(total[4])
        else:
            valeur.append(2)
            card2.append(total[1])
            del card[max_key]
            max_key, max_same = max(card, key=card.get), card[max(card, key=card.get)]
            
            card3.append(total[2])
            card4.append(total[3])
            card5.append(total[4])
    else:
#         print(card)
        valeur.append(1)
        card_1.append(total[0])
        del card[max_key]
        max_key, max_same = max(card, key=card.get), card[max(card, key=card.get)]
        card2.append(total[1])
        del card[max_key]
        max_key, max_same = max(card, key=card.get), card[max(card, key=card.get)]
        card3.append(total[2])
        del card[max_key]
        max_key, max_same = max(card, key=card.get), card[max(card, key=card.get)]
        card4.append(total[3])
        del card[max_key]
        max_key, max_same = max(card, key=card.get), card[max(card, key=card.get)]
        card5.append(total[4])
#     df['cards'][i]=card
    i+=1
df['game']=game
df['valeur']=valeur
df['card_1']=card_1
df['card2']=card2
df['card3']=card3
df['card4']=card4
df['card5']=card5
```
