```python
import pandas as pd
df = pd.read_csv(
            r'C:\Users\SA012547\Documents\advent_of_code_2023\day4.txt',sep=':', encoding='latin-1', header=None) #['nb_game', 'liste_cube'])
```

```python
df.columns=['card', 'nb']
df[['win', 'mine']] = df['nb'].str.split('|', n=1, expand=True)
```

```python
df['win']= df['win'].str.split(' ')
df['mine']= df['mine'].str.split(' ')
```

```python
import numpy as np
df['intersection'] = [list(set(a).intersection(set(b))) 
                      for a, b in zip(df.win, df.mine)]
df['len']=(df['intersection'].str.len()-1)
df['point']=np.where(df['len'] == 0,0, 2**(df['len'])/2)  # /2 car 2**df['len'] pose souci avec -1? meme avec le np.where
df
```
