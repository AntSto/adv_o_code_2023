```python
import pandas as pd
data_1 = pd.read_csv(
            r'C:\Users\SA012547\Documents\advent_of_code_2023\day_2.txt',sep=':', encoding='latin-1', header=None) #['nb_game', 'liste_cube'])
```

```python
# 12 red cubes, 13 green cubes, and 14 blue cubes. What is the sum of the IDs of those games?
```

```python
data_1.columns = ['nb_game', 'liste_cube']
```

```python
data_1['red']=data_1['liste_cube'].replace(r'\d+ (blue|green),*', '', regex=True)
# data_1['red1']=data_1['red'].replace(r'(;)', ',', regex=True)
data_1['red2']=data_1['red'].replace(r'(red|;)', '', regex=True)
data_1['red2']=data_1['red2'].replace(r'( *, *)+', ',', regex=True)
data_1['red3']=data_1['red2'].str.split(',')
# data_1['max_red']=data_1['red3'].astype(int)
data_1
```

```python
data_1['green1']=data_1['liste_cube'].replace(r'\d+ (blue|red),*', '', regex=True) 
data_1['green1']=data_1['green1'].replace(r'( *|,*|green)','', regex=True)
# data_1['redx']=data_1['redx'].replace(r'(^ *,|, *$| red)','', regex=True)
data_1['green1']=data_1['green1'].replace(r';|,',',', regex=True)
data_1['green1']=data_1['green1'].replace(r',+',',', regex=True)
data_1['green1']=data_1['green1'].replace(r'(^,|,$)', '', regex=True)
data_1['green1']=data_1['green1'].str.split(',')
# data_1['redx3'] =[[int(idx, 10) for idx in x.split(",")] for x in data_1['redx']]
data_1['green_max']=data_1['green1'].explode().astype(int).max(level=0)
```

```python
data_1['blue']=data_1['liste_cube'].replace(r'\d+ (green|red),*', '', regex=True) 
data_1['blue']=data_1['blue'].replace(r'( *|,*|blue)','', regex=True)
# data_1['redx']=data_1['redx'].replace(r'(^ *,|, *$| red)','', regex=True)
data_1['blue']=data_1['blue'].replace(r';|,',',', regex=True)
data_1['blue']=data_1['blue'].replace(r',+',',', regex=True)
data_1['blue']=data_1['blue'].replace(r'(^,|,$)', '', regex=True)
data_1['blue']=data_1['blue'].str.split(',')
# data_1['redx3'] =[[int(idx, 10) for idx in x.split(",")] for x in data_1['redx']]
data_1['blue_max']=data_1['blue'].explode().astype(int).max(level=0)
```

```python
data_1['redx']=data_1['liste_cube'].replace(r'\d+ (blue|green),*', '', regex=True) 
data_1['redx']=data_1['redx'].replace(r'( *|,*|red)','', regex=True)
# data_1['redx']=data_1['redx'].replace(r'(^ *,|, *$| red)','', regex=True)
data_1['redx']=data_1['redx'].replace(r';|,',',', regex=True)
data_1['redx']=data_1['redx'].replace(r',+',',', regex=True)
data_1['redx']=data_1['redx'].replace(r'(^,|,$)', '', regex=True)
data_1['redx2']=data_1['redx'].str.split(',')
# data_1['redx3'] =[[int(idx, 10) for idx in x.split(",")] for x in data_1['redx']]
data_1['red_max']=data_1['redx2'].explode().astype(int).max(level=0)
# data_1['redx3']=data_1['redx2'].apply(lambda x: x.astype(int))
data_1
```

```python
# 12 red cubes, 13 green cubes, and 14 blue cubes. What is the sum of the IDs of those games?
```

```python
data_2 = data_1[(data_1.red_max <13) &(data_1.green_max <14)&(data_1.blue_max <15)]
data_2['nb_game2']=data_2['nb_game'].replace(r'Game', '', regex=True).astype(int) 
sum(data_2['nb_game2'])
```
