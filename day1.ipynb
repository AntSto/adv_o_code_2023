```python
import pandas as pd
data_1 = pd.read_csv(
            r'C:\Users\SA012547\Documents\advent_of_code_2023\advent_of_code.csv',sep=';', encoding='latin-1', header=None)
```

```python
dict_replace_0 = {
    '(twone)':'two one',
    '(oneight)' : ' one eight',
    '(threeight)' : 'three eight',
    'fiveight': 'five eight',
    'nineight' : 'nine eight',
    'eightwo' : 'eight two',
    'eighthree': 'eight three',
    'seven nine': 'seven nine'}

dict_replace_1 = {
    '(one)':'1',
    '(two)':'2',
    '(three)':'3',
    '(four)':'4',
    '(five)':'5',
    '(six)':'6',
    '(seven)':'7',
    '(eight)':'8',
    '(nine)':'9'
}
```

```python
data_1['string0'] = data_1
```

```python
data_1['string00'] = data_1['string0'].replace(dict_replace_0, regex=True)
data_1['string'] = data_1['string00'].replace(dict_replace_1, regex=True)
```

```python
data_1['number1'] = data_1['string'].str.extract('(\d{1})')
```

```python
data_1['number2'] = data_1['string'].str[::-1].str.extract('(\d{1})')
```

```python
data_1['full_num'] = pd.to_numeric(data_1['number1'] + data_1['number2'])
```

```python
data_1['sum'] =  data_1["full_num"].sum()
```
