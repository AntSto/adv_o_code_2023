```python
import pandas as pd
df = pd.read_csv(
            r'day9.txt',sep=',', encoding='latin-1', header=None) #['nb_game', 'liste_cube'])
```

```python
df.columns=['info']
# df['info']=df['info'].str.replace(' +', ' ', regex=True).str.split(' ')
# df['info']=[int(x) for x in df['info']]
 # [[int(idx, 16) for idx in x.split(",")] for x in df['rnti']]
df['info']=[[int(idx) for idx in x.replace(' +', ' ').split(' ')] for x in df['info']]
df['dif']=[[y-x for x,y in zip(z, z[1:])] for z in df['info']]
df.head()
```

```python
df['last']=0
sum_col=[]
prev_col=[[] for x in df['info']]
first=[]
res_col = [[] for x in df['info']]
dif= []
i=0
while i < len(df['info']):
    y=0
    lst = df['info'][i]
    while y < len(lst):
        res = [int(y)-int(x) for x, y in zip(lst, lst[1:])]
        res_col[i].append(res)
        lst=res
    sum_all=sum([sum(x[-1:]) for x in res_col[i]])
    sum_col.append(sum_all) 
    z = 0
    
    i+=1
    
df['res']=res_col
df['sum']=sum_col
df['final']=df['info'].str.get(-1)+df['sum']
df.head(2)
```
